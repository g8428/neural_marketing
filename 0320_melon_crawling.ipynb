```python
from selenium.common.exceptions import NoSuchFrameException

for i in range(1,5):
    driver.get('https://www.melon.com/dj/powerdj/djpowerdj_list.htm')
    driver.implicitly_wait(3) #2초 대기
    driver.find_element_by_xpath(f'/html/body/div/div[3]/div/div/div[3]/div[2]/div[{i}]/div[2]/h3/a').click()
    driver.find_element_by_xpath('/html/body/div/div[2]/div/div/div[1]/ul/li[4]/a/span').click()
    for j in range(1,20):
        driver.implicitly_wait(3) #2초 대기
        driver.find_element_by_xpath(f'/html/body/div/div[2]/div/div/div[3]/div/div/table/tbody/tr[{j}]/td[2]/div/div/dl/dt/a').click()
        driver.implicitly_wait(4) #4초 대기
        tag = driver.find_element_by_xpath(f"""//*[@id="conts"]/div[2]/div/div[2]/div[3]""")
        print(tag.text)
        for l in range(1,11):
            driver.implicitly_wait(4) #4초 대기
            songs = driver.find_element_by_xpath(f"""//*[@id="frm"]/div/table/tbody/tr[{l}]/td[5]/div/div/div[1]/span/a""")
            print(songs.text)
        driver.back()
```
